```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=';')
df.describe()
```

### Aggregate of a column  - sum, mean, median etc 

```python
# Either with dot or brackets we can call a column 
print(df.sulphates.sum())
print(df['sulphates'].sum())
```

### Unique values of a column

```python
print(f"Unique values of the quality column: {df['quality'].unique()}")
```

### Group by 1 or multiple columns 

```python
### Group by one column
df.groupby("quality").mean()
```

```python
### group by multiple columns 
df.groupby(["quality", "residual sugar"]).mean()
```

```python
# calculate the mean of a specific column group by one specific column
df.groupby(["quality"])["alcohol"].mean()
```

### Group by a speficic column but not use its groups as index 

```python
df.groupby("quality", as_index=False).mean()
```

### Get the describe function for all the groups  

```python
grouped = df.groupby("quality")
grouped.describe()
```

### Descibe of only one specific column 

```python
grouped.describe()["fixed acidity"]
```

### Compute summary statistics for numeric variables

```python
df.describe(include = [np.number])
```

### Describe and percentiles

```python
df.describe(percentiles = [0.01, .1, .9, 0.99])
```

### All the invividual groups of the grouped object 

```python
grouped.groups.keys() 
```

### Get back all the rows for this specific group 

```python
df.loc[grouped.groups[3]] 
```

### Group by & Aggregation: Multiple Methods with a groupby

```python
df.groupby("quality").agg({"fixed acidity": ["sum", "mean", "std"]})
```

```python
#  flatten the index into one level
df.columns = df.columns.to_flat_index()
df
```

```python
# different aggregation functions to different columns 
df.groupby("quality").agg({
    "fixed acidity": "sum",
    "volatile acidity": ["mean", "std"]
})
```

### Named aggregations 

```python
df.groupby("quality").agg(
    min_fa=("fixed acidity", "min"),
    var_va=("volatile acidity", "var")
)
```

```python
df.groupby('quality').agg(
    sum_alcohol = ('alcohol', 'sum'),
    geomean_of_hue = ('pH', lambda x: x.prod() ** (1 / len(x)))
)
```

```python
def field_range(s):
    return max(s) - min(s)

df.groupby("quality")["fixed acidity"].agg(min_fixed_acid = "min", max_fixed_acid = "max", max_minus_min= field_range)
```

```python
grouped = df.groupby("quality").agg({
    "fixed acidity": "sum",
    "volatile acidity": ["mean", "std"]
}).reset_index()
grouped
```

### Drop NA

```python
df.dropna(how='all') # Drop only if ALL columns are NaN
df.dropna(thresh=2)  # Drop row if it does not have at least two values that are **not** NaN
df.dropna(subset=["pH"]) # Drop only if NaN in specific column
```
