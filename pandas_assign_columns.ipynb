```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=';')
df.describe()
```

### Assign new columns & lamda function

```python

df1 = df.assign(testcol1=lambda df: df.pH/2, testcol2=lambda df: df.pH/2)
df1
```

### Modify a single column & apply function

```python
# Apply function
def add_3(x):
   return x+3
df1.pH = df1.pH.apply(add_3)
df1
```

### Modify a multiple columns & apply function

```python
df1[['pH','testcol1']] = df1[['pH','testcol1']].apply(add_3)
df1
```


### Modify a multiple columns & lambda function

```python
df1[['pH','testcol1']] = df1[['pH','testcol1']].apply(lambda x : x+9000, axis=1)
df1
```

### Conditionally modify a column 

```python
df1.testcol2 = df1.testcol2.apply(lambda x : x + 100 if x > 9.2 else x-100)
df1
```

### Create a new column with List comprehension

```python
df1['squared_density'] = [i**2 for i in df1['density']]
df1
```

###  Assign values on a subset of a column

```python
df1.loc[df1.chlorides>0.075, 'pH'] = df1.loc[df1.chlorides>0.075, 'pH'] / 10
df1
```
